```python
import pandas as pd
import csv
import os
```

```python
health_df = pd.read_csv("data.csv")
health_df
```

```python
new_df = health_df.loc[:, ['Country Name', 'Indicator Name','Indicator Code', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
new_df
```

```python
# Group the dataframe by 'Country Name' and sum the values in each row
indicator_names = new_df['Indicator Name'].unique()
indicator_names
```

```python
indicator_name = ['Birth rate, crude (per 1,000 people)', 'Death rate, crude (per 1,000 people)', 'Population, total', 'Population growth (annual %)']
columns = ['Country Name', 'Indicator Name','2006','2007','2008','2009','2010','2011','2012', '2013','2014']
countries = ['Germany', 'Turkey']
df_filtered = health_df[(health_df['Indicator Name'].isin(indicator_name)) 
                        & (health_df['Country Name'].isin(countries))][columns]
df_filtered = df_filtered.round(2)
df_filtered
```

```python
import matplotlib.pyplot as plt
```

```python
birth_rates = df_filtered[df_filtered['Indicator Name'] == 'Birth rate, crude (per 1,000 people)']
birth_rates
```

```python
birth_rates = birth_rates.drop('Indicator Name', axis=1)
birth_rates
```

```python
birth_rates
```

```python
years = [x for x in range(2006,2015)]
germany_b = []
turkey_b = []
for year in years:
    germany_b.append(birth_rates[str(year)][39053])
    turkey_b.append(birth_rates[str(year)][82523])
```

```python
plt.plot(years, germany_b, label="Germany")
plt.plot(years, turkey_b, label="Turkey")
plt.legend()
plt.title('Birth Rates')
plt.show()
```

```python
indicator_name = ['Female population 20-24',
'Male population 20-24']
columns = ['Country Name', 'Indicator Name','2006','2007','2008','2009','2010','2011','2012', '2013','2014','2015']
countries = ['Germany', 'Turkey']
df_ea = health_df[(health_df['Indicator Name'].isin(indicator_name)) 
                        & (health_df['Country Name'].isin(countries))][columns]
df_ea
```

```python
ger_f_ea = df_ea[(df_ea['Country Name'] == 'Germany') & (df_ea['Indicator Name'] == 'Female population 20-24')].values.flatten()[2:]
ger_m_ea = df_ea[(df_ea['Country Name'] == 'Germany') & (df_ea['Indicator Name'] == 'Male population 20-24')].values.flatten()[2:]
tur_f_ea = df_ea[(df_ea['Country Name'] == 'Turkey') & (df_ea['Indicator Name'] == 'Female population 20-24')].values.flatten()[2:]
tur_m_ea = df_ea[(df_ea['Country Name'] == 'Turkey') & (df_ea['Indicator Name'] == 'Male population 20-24')].values.flatten()[2:]

years = [x for x in range(2006,2015)]
ger_f_ea = []
ger_m_ea = []
tur_f_ea = []
tur_m_ea = []

for year in years:
    ger_f_ea.append(df_ea[str(year)][39089])
    ger_m_ea.append(df_ea[str(year)][39147])
    tur_f_ea.append(df_ea[str(year)][82559])
    tur_m_ea.append(df_ea[str(year)][82617])

plt.plot(years, ger_f_ea, label='Germany - Female')
plt.plot(years, ger_m_ea, label='Germany - Male')
plt.plot(years, tur_f_ea, label='Turkey - Female')
plt.plot(years, tur_m_ea, label='Turkey - Male')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of 20-24 years old by Gender')
plt.legend()

plt.show()
```

```python
indicator_name = ['Incidence of tuberculosis (per 100,000 people)','Tuberculosis death rate (per 100,000 people)']
columns = ['Country Name', 'Indicator Name','2006','2007','2008','2009','2010','2011','2012', '2013','2014']
countries = ['Germany', 'Turkey',]
df_tb = health_df[(health_df['Indicator Name'].isin(indicator_name)) 
                        & (health_df['Country Name'].isin(countries))][columns]
df_tb
```

```python
gtbi = df_tb[(df_tb['Country Name'] == 'Germany') & (df_tb['Indicator Name'] == 'Incidence of tuberculosis (per 100,000 people)')].values.flatten()[2:]
gtbd = df_tb[(df_tb['Country Name'] == 'Germany') & (df_tb['Indicator Name'] == 'Tuberculosis death rate (per 100,000 people)')].values.flatten()[2:]
ttbi = df_tb[(df_tb['Country Name'] == 'Turkey') & (df_tb['Indicator Name'] == 'Incidence of tuberculosis (per 100,000 people')].values.flatten()[2:]
ttbd = df_tb[(df_tb['Country Name'] == 'Turkey') & (df_tb['Indicator Name'] == 'Tuberculosis death rate (per 100,000 people')].values.flatten()[2:]

years = [x for x in range(2006,2015)]
gtbi = []
gtbd = []
ttbi = []
ttbd = []
for year in years:
    gtbi.append(df_tb[str(year)][39127])
    gtbd.append(df_tb[str(year)][39314])
    ttbi.append(df_tb[str(year)][82597])
    ttbd.append(df_tb[str(year)][82784])

plt.plot(years, gtbi, label='Germany - Incidence of TB per 100,000 people')
plt.plot(years, gtbd, label='Germany - TB Deaths per 100,000 people')
plt.plot(years, ttbi, label='Turkey - Incidence of TB per 100,000 people')
plt.plot(years, ttbd, label='Turkey - TB Deaths per 100,000 people')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('TB incidences and deaths per 100,000 people')
plt.legend()

plt.show()
```
