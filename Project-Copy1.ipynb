```python
import pandas as pd
pd.set_option('display.max_rows', 500)
import matplotlib.pyplot as plt
```

```python
path='data.csv'
```

```python
health_df=pd.read_csv(path)
```

```python
health_df.drop(health_df.columns[4:49], axis=1, inplace=True)
```

```python
health_df
```

```python
country=['Turkey', 'Germany']
```

```python
country_data=health_df[health_df['Country Name'].isin(country)]
columns=['Country Name','Indicator Name','2005','2006', '2007','2008','2009','2010','2011','2012','2013','2014','2015']
country_data=country_data[columns]
```

```python
country_data
```

```python
indicator_name = ['Health expenditure, private (% of GDP)','Health expenditure, public (% of government expenditure)']
health_care_df = country_data[country_data['Indicator Name'].isin(indicator_name)]

health_care_df
```

```python
graph_1=health_care_df.drop(['Indicator Name', '2015'],axis=1).set_index(['Country Name']).transpose()
graph_1.columns=['Germany Total Private Spending (% GPD)', 'Germany Total Public Spending (% GPD)', 'Turkey Total Private Spending (% GPD)', 'Turkey Total Public Spending (% GPD)']
graph_1
```

```python
graph_1.plot(kind='line');
```

```python
indicator_name_2=['Female population 20-24', 'Female population 25-29']
female_pop_20_29=country_data[country_data['Indicator Name'].isin(indicator_name_2)]
female_pop_20_29
```

```python
graph_2=female_pop_20_29.drop(['Indicator Name', '2015'],axis=1).set_index(['Country Name']).transpose()

graph_2.columns=['Germany Females 20-24', 'Germany Females 25-29', 'Turkey Females 20-24', 'Turkey Females 25-29']
graph_2
```

```python
graph_2.plot(kind='line');
```
